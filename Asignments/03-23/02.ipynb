Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

Note: Also add title, xlabel, and ylabel to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x,y)
plt.title("My Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:

```python
import numpy as np

# Line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

subplot function is used to plot multiple sets of data in a single figure in order to make it easier to do comparison between them.

```python
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))

A = axes[0, 0]
A.plot(x1, y1)
A.set_title('A')

B = axes[0, 1]
B.plot(x2, y2)
B.set_title('B')

C = axes[1, 0]
C.plot(x3, y3)
C.set_title('C')

D = axes[1, 1]
D.plot(x4, y4)
D.set_title('D')
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.


```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot([box1, box2])
plt.show()
```

Box plot is the graphical representaion of dataset to display the distribution through five key summary statistics:
- minimum
- 25 %ile
- 50 %ile
- 75 %ile
- maximum

It provides insights into the centeral tendency, spread and presence of outliers in data.
