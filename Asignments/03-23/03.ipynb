Q1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Ans: Seaborn plots:

1. **box plot**: Display the distribution of data based on 5 statistical summaries (min, 3-quartiles, max). `sns.boxplot()`
1. **barplot**: Display categorical data with rectangular bars. Height of bars determine the size of category. `sns.barplot()`
1. **histogram**: Display the distribution of a continuous variable. Helps understanding the frequency distribution of data. `sns.histplot`
1. **scatter plot**: Display the relationship between two continuous variables, useful for identifying patterns, trends or correlations.`sns.histplot()`
1. **Heat map**: visualize the matrix of data with color coded cells to reveal the patterns or correlations. `sns.heatmap()`

Q2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns

fmri = sns.load_dataset('fmri')
sns.lineplot(data=fmri, x='timepoint', y='signal', style='event', hue='region')
```

Q3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x = 'pclass', y = 'age' and y = 'fare'.

```python
tit = sns.load_dataset('titanic')
```

```python
sns.boxplot(data=tit, x='pclass', y='age')
```

```python
sns.boxplot(data=tit, x='pclass', y='fare')
```

Q4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.

```python
diamonds = sns.load_dataset('diamonds')
```

```python
sns.histplot(data=diamonds, x='price', hue='cut')
```

Q5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.


```python
iris = sns.load_dataset('iris')
```
