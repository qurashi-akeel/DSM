Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a scatter plot for age and fare columns in the titanic dataset.

```python
import seaborn as sns
import plotly.graph_objects as go

titanic = sns.load_dataset('titanic')
```

```python
fig1 = go.Figure()

fig1.add_trace(
    go.Scatter(x=titanic['age'], y=titanic['fare'], mode='markers')
)
```

Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

```python
import plotly.express as px

df = px.data.tips()
fig2 = px.box(df['tip'], df['total_bill'])
fig2.show()
```

Q3: Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.

```python
df = px.data.tips()

fig3 = px.histogram(df, x="sex", y="total_bill", pattern_shape="smoker", color="day")
fig3.show()
```

Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.

```python
df = px.data.iris()

fig4 = px.scatter(df, x="sepal_width", y="sepal_length", color="species")
fig4.show()
```
