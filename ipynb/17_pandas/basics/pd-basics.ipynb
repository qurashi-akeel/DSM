```python
csv_url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df_ol = pd.read_csv(csv_url) # dataframe from url
```

```python
df_ol
```

```python
# Save the data for future
df_ol.to_csv('titanic.csv', index=False)
```

> This can be also used to store the data to a file after operations/manipulations. we can choose the file format as:
1. csv - `df_ol.to_csv('titanic.csv')`
2. json - `df_ol.to_json('titanic.json')`
3. parquet - `df_ol.to_parquet('titanic.parquet')`
4. excel - `df_ol.to_excel('titanic.xlsx')` etc...
