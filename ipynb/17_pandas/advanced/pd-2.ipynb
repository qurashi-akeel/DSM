```python
df['Fare_INR'] = df['Fare'].apply(lambda x: (x * 80))
df.head()
```

```python
df['name_len'] = df['Name'].apply(len)
df.head()[['Name', 'name_len']]
```

```python
def create_class(fare):
    match fare:
        case fare if fare <= 0:
            return "f"
        case fare if fare <= 15:
            return "e"
        case fare if fare < 50:
            return "d"
        case fare if fare < 80:
            return "c"
        case fare if fare < 130:
            return "b"
        case fare if fare < 220:
            return "a"
        case fare if fare < 270:
            return "a+"
        case _:
            return "a++"
        
df['PGrade'] = df['Fare'].apply(create_class)
```

```python
grade = df['PGrade'] == 'a'
survived = len(df.loc[(df['Survived'] == 1) & (grade)])
total = len(df.loc[(grade)])

if total > 0:
    print(
        "Total: ", total,
        "\nSurvived: ", survived,
        "\nDied: ", total - survived,
        "\nPercentage :",survived / total * 100
    )
else:
    print("Total: ", total)
```
