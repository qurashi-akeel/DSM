```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(20)
y = np.random.rand(20)
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y, c='red', alpha=0.4, marker='+')
plt.title("My Scatter Plot")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid()
```

```python
x = np.linspace(1, 10, 50) # 50 elements between 1 - 10 (with uniform gap)
y = np.sin(x)
```

```python
plt.figure(figsize=(5,2))
plt.plot(x,y, '--g')
plt.title("Sin Wave Graph")
```

```python
x = ['a', 'b', 'c', 'd', 'e']
y = np.random.randint(2, 10, 5)
```
