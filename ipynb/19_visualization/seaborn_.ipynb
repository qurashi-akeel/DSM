```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x=iris.sepal_length, y=iris.sepal_width, hue=iris.species)
```

```python
sns.scatterplot(x=iris.sepal_length, y=iris.petal_length,  hue=iris.species)
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.scatterplot(x=tips.total_bill, y=tips.tip, hue=tips.day)
```

```python
tips.smoker.value_counts()
```

```python
sns.relplot(data=tips, x=tips.total_bill, y=tips.tip, hue=tips.smoker, aspect=3, height=7)
```

```python
sns.relplot(
    data=tips,
    x=tips.total_bill,
    y=tips.tip,
    style='size',
    hue="time",
    aspect=2,
    height=7
)
```

```python
sns.catplot(x='day', y='total_bill', data=tips)
```

```python
sns.catplot(x='smoker', y='tip', data=tips)
```

```python
sns.jointplot(data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.pairplot(data=iris)
```
